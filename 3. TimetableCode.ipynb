```
count = 0
data_240101 = data[data["stop_no"] == 240101]

for date in data_240101.date.unique():
    abs_pointer = 0

    by_date = data_240101[data_240101.date == date]
    for time in by_date.query_time.unique():
        test = by_date[by_date.query_time == time]

        

        day = pd.to_datetime(date)
        current_pointer = -1
        extra = 0

        timetable = weekday_times[240101]



        if day.weekday() == 5:
            timetable = saturday_times[240101]
        if day.weekday() == 6:
            timetable = sunday_times[240101]

        for index in test.index.unique():
            pointer = False
            query = data_240101.at[index,"query_time"]
            arrival = data_240101.at[index, "expected_arrival"]
            if arrival.hour < 1:
                day = day - timedelta(days=1)
                old_time = arrival
                arrival = datetime(1,1,1,23,59).time()
                if day.weekday() == 5:
                    timetable = saturday_times[240101]
                if day.weekday() == 6 or day.date() in b_h_objects:
                    timetable = sunday_times[240101]
                extra = (old_time.hour*60 + old_time.minute) + 1
                data.at[index, "date"] = day
            last_time = pd.to_datetime(timetable[-1]).time()
            last_difference = (arrival.hour*60 + arrival.minute) - (last_time.hour*60 + last_time.minute)
            lowest = 1441
            nearest = 0
            for j in range(len(timetable)-1):
                time = pd.to_datetime(timetable[j]).time()
                nxt = pd.to_datetime(timetable[j+1]).time()                        
                distance = (nxt.hour*60 + nxt.minute) - (time.hour*60 + time.minute)
                difference = (arrival.hour*60 + arrival.minute) - (time.hour*60 + time.minute)

                if -5 <= difference < (distance-5):
                    current_pointer = j
                    pointer = True
                    break
                elif abs(difference) < lowest:
                    lowest = abs(difference) 
                    nearest = j
            if not pointer:          
                current_pointer = nearest
                if (abs(last_difference) < lowest):
                    current_pointer = len(timetable)-1
            if current_pointer >= len(timetable):
                  current_pointer = len(timetable)-1
            time = pd.to_datetime(timetable[current_pointer]).time()
            difference = (arrival.hour*60 + arrival.minute) - (time.hour*60 + time.minute) + extra
            data.at[index, "timetable"] = time
            data.at[index, "wait_time"] = difference

data.to_csv("/content/drive/My Drive/vmware/Project/better_timetable.csv", sep=",")
```
